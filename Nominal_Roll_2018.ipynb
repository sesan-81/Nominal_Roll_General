```python
def calculate_age(dob):
    today = datetime.date.today()
    age = today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))
    return age
```

```python
df.columns 
```

```python
def calculate_retirement(appointment_date):
    #age_at_retirement = 60
    service_year=35
    appointment_year = appointment_date.year
    retirement_year = appointment_year + service_year
    retirement_month = appointment_date.month
    retirement_day = appointment_date.day
    retirement_date = datetime.date(retirement_year, retirement_month, retirement_day)
    return retirement_date
```

```python
l1=[]
for line in df['DATE OF ENLIST. ']:
    appointment_date = line
    l1.append(calculate_retirement(appointment_date))
```

```python
df['retirement_date']=l1
```

```python
l2=[]
for line in df['date_of_birth']:
    birth_date = line
    l2.append(calculate_age(birth_date))
```

```python
df['age'] = l2
```
